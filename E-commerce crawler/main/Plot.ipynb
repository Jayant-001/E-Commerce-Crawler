```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

data = {'amazon': 2300, 'Flipkart': 2200, 'olx': 1200}
name = list(data.keys())
price = list(data.values())
plt.figure(figsize=(8,5))
plt.bar(name, price, color='brown', width=0.5)
plt.xlabel('Websites offers product')
plt.ylabel('Price')
plt.title('Redmi k20 pro')
plt.show()
```

```python
d1 = [10000, 9800, 7000]
d2 = [9500, 9500, 7000]
x = np.arange(3)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x+0.0, d1, color='b', width=0.25)
ax.bar(x+0.25, d2, color='r', width=0.25)
ax.legend(labels=['initial price', 'discount price'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'sites': ['Amazon', 'Flipkart', 'OLX'],
    'current': [1200, 1150, 900],
    'offer': [1000, 1000, 850],
}
df = pd.DataFrame(data, columns = ['name', 'current', 'offer'])
pos = [1,2,3]
width = 0.25

fig, ax = plt.subplots(figsize=(10,5))

plt.bar(pos[0], df['current'], width)
plt.bar(pos[1] + 0.5, df['offer'], width)
plt.show()
```
